# import libraries

```python
import pandas as pd
import numpy as np
```

# read dataset

```python
df = pd.read_csv('iris_dataset.csv')
df.shape
```

# find k=3 nearest neighbour

```python
train_df = df.iloc[:-1]
test_df = df.iloc[-1]
print(f"train_df.shape: {train_df.shape}")
print(f"test_df.shape: {test_df.shape}")

train = train_df.drop(columns=["target"]).to_numpy()
test = test_df.drop(labels=["target"]).to_numpy()
test_class = test_df['target']
print(f"train.shape: {train.shape}")
print(f"test.shape: {test.shape}")

distance = np.linalg.norm(train - test, axis=1)
print(f"distance.shape: {distance.shape}, distance[0]: {distance[0]}")
```

```python
idx = np.argsort(distance)[:3]
nearest_neighbours_classes = train_df.iloc[idx]['target']
values, counts = np.unique(nearest_neighbours_classes, return_counts=True)
most_frequent = values[np.argmax(counts)]
most_frequent
```

```python
print(f"most_frequent ({most_frequent}) == item_class ({test_class}): {most_frequent == test_class}")
```

# do normalization

```python
df_norm = df.copy()
numeric_cols = df.columns.difference(["target"])
df_norm[numeric_cols] = \
    df[numeric_cols].transform(lambda x: (x - x.mean()) / x.std())

df_norm.head(1).T
```

# find k=3 nearest neighbours

```python
train_df = df_norm.iloc[:-1]
test_df = df.iloc[-1]
print(f"train_df.shape: {train_df.shape}")
print(f"test_df.shape: {test_df.shape}")

train = train_df.drop(columns=["target"]).to_numpy()
test = test_df.drop(labels=["target"]).to_numpy()
test_class = test_df['target']
print(f"train.shape: {train.shape}")
print(f"test.shape: {test.shape}")

distance = np.linalg.norm(train - test, axis=1)
print(f"distance.shape: {distance.shape}, distance[0]: {distance[0]}")

idx = np.argsort(distance)[:3]
nearest_neighbours_classes = train_df.iloc[idx]['target']
values, counts = np.unique(nearest_neighbours_classes, return_counts=True)
most_frequent = values[np.argmax(counts)]

print(f"most_frequent ({most_frequent}) == item_class ({test_class}): {most_frequent == test_class}")
```
